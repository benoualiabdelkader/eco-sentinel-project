## Integrate Script Execution

### Subtask:
Define Python functions that will be called when the buttons are pressed, executing the corresponding project scripts using subprocess calls and capturing their output.


**Reasoning**:
To integrate script execution into the GUI, I will modify the `create_gui()` function to include `subprocess` calls within the button command functions. This will allow the buttons to execute the specified Python scripts and capture their output for display in the `output_text` widget.



````python
def create_gui():
    import tkinter as tk
    from tkinter import scrolledtext
    import subprocess
    import os # Import os for path manipulation

    root = tk.Tk()
    root.title("Project Script Runner")
    root.geometry("800x600")

    # --- Configuration for local execution ---
    # IMPORTANT: When running locally, ensure 'project_base_path' points to the directory
    # containing your 'project' folder. For this example, we assume '/content/project'
    # is accessible, but for local execution, you might need something like:
    # project_base_path = os.path.join(os.getcwd(), 'project') # if 'project' is in current dir
    # Or a fixed path:
    # project_base_path = '/path/to/your/local/project_directory'

    # For this Colab-generated script, we'll use the *intended* Colab path for clarity
    # but remind the user to adjust it for local setup.
    project_base_path = '/content/project'

    # Frame for buttons
    button_frame = tk.Frame(root)
    button_frame.pack(pady=10)

    # Output display area
    output_text = scrolledtext.ScrolledText(root, wrap=tk.WORD, width=90, height=25)
    output_text.pack(padx=10, pady=10, fill=tk.BOTH, expand=True)
    output_text.insert(tk.END, "GUI initialized. Click a button to run a script.\n")

    def run_script(script_name):
        output_text.delete(1.0, tk.END) # Clear previous output
        output_text.insert(tk.END, f"Running {script_name}...\n")
        output_text.see(tk.END)

        script_path = os.path.join(project_base_path, script_name)

        try:
            # Execute the script using subprocess
            # We use `python` explicitly to ensure the correct interpreter is used
            result = subprocess.run(['python', script_path],
                                    capture_output=True, text=True, check=True)

            if result.stdout:
                output_text.insert(tk.END, "\n--- STDOUT ---\n")
                output_text.insert(tk.END, result.stdout)
            if result.stderr:
                output_text.insert(tk.END, "\n--- STDERR ---\n")
                output_text.insert(tk.END, result.stderr)
            output_text.insert(tk.END, f"\n{script_name} finished successfully.\n")

        except subprocess.CalledProcessError as e:
            output_text.insert(tk.END, f"\nError running {script_name}:\n")
            output_text.insert(tk.END, f"Command: {' '.join(e.cmd)}\n")
            output_text.insert(tk.END, f"Return Code: {e.returncode}\n")
            output_text.insert(tk.END, f"STDOUT: {e.stdout}\n")
            output_text.insert(tk.END, f"STDERR: {e.stderr}\n")
        except FileNotFoundError:
            output_text.insert(tk.END, f"\nError: Python interpreter or script '{script_path}' not found.\n")
            output_text.insert(tk.END, "Please ensure Python is in your PATH and the script path is correct.\n")
        except Exception as e:
            output_text.insert(tk.END, f"\nAn unexpected error occurred: {e}\n")

        output_text.see(tk.END) # Scroll to the end of the output

    # Add buttons
    btn_train_and_evaluate = tk.Button(button_frame,
                                       text="Run train_and_evaluate.py",
                                       command=lambda: run_script('train_and_evaluate.py'))
    btn_train_and_evaluate.pack(side=tk.LEFT, padx=5)

    btn_train_eco_models = tk.Button(button_frame,
                                     text="Run train_eco_models.py",
                                     command=lambda: run_script('train_eco_models.py'))
    btn_train_eco_models.pack(side=tk.LEFT, padx=5)

    root.mainloop()

print("GUI function 'create_gui()' updated with script execution logic.")
print("This function should be called from a local Python environment, after adjusting the 'project_base_path'.")
````

            output_text.insert(tk.END, f"Changed working directory to: {os.getcwd()}\n")

            # Execute the script using subprocess
            # We use `python` explicitly to ensure the correct interpreter is used
            result = subprocess.run(['python', script_full_path],
                                    capture_output=True, text=True, check=True)

            if result.stdout:
                output_text.insert(tk.END, "\n--- STDOUT ---\n")
                output_text.insert(tk.END, result.stdout)
            if result.stderr:
                output_text.insert(tk.END, "\n--- STDERR ---\n")
                output_text.insert(tk.END, result.stderr)
            output_text.insert(tk.END, f"\n{script_name} finished successfully.\n")

        except subprocess.CalledProcessError as e:
            output_text.insert(tk.END, f"\nError running {script_name}:\n")
            output_text.insert(tk.END, f"Command: {' '.join(e.cmd)}\n")
            output_text.insert(tk.END, f"Return Code: {e.returncode}\n")
            output_text.insert(tk.END, f"STDOUT: {e.stdout}\n")
            output_text.insert(tk.END, f"STDERR: {e.stderr}\n")
        except FileNotFoundError as e:
            output_text.insert(tk.END, f"\nError: Python interpreter or script '{script_full_path}' not found.\n")
            output_text.insert(tk.END, f"Please ensure Python is in your PATH and the script path is correct. Details: {e}\n")
        except Exception as e:
            output_text.insert(tk.END, f"\nAn unexpected error occurred: {e}\n")
        finally:
            # Always change back to the original working directory
            os.chdir(original_cwd)
            output_text.insert(tk.END, f"Reverted working directory to: {os.getcwd()}\n")
            
        output_text.see(tk.END) # Scroll to the end of the output

    # Add buttons
    btn_train_and_evaluate = tk.Button(button_frame,
                                       text="Run train_and_evaluate.py",
                                       command=lambda: run_script('train_and_evaluate.py'))
    btn_train_and_evaluate.pack(side=tk.LEFT, padx=5)

    btn_train_eco_models = tk.Button(button_frame,
                                     text="Run train_eco_models.py",
                                     command=lambda: run_script('train_eco_models.py'))
    btn_train_eco_models.pack(side=tk.LEFT, padx=5)

    root.mainloop()

if __name__ == '__main__':
    create_gui()
